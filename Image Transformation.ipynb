# 画像変換 ~ tf.imageライブラリ紹介 ~

## Encoding and Decoding
### Encoding
文字型の0次元Tensorを、デコードしてN次元のTensorに変換する
- tf.image.decode_gif() # アニメーションGIFにも対応っぽい?
- tf.image.decode_jpeg()
- tf.image.decode_png()

### Decoding
N次元のTensorを変換して文字列にする(バイト配列)
- tf.image.encode_jpeg()
- tf.image.encode_png()

## Resizing
### 注意
**Resizeする時は事前にfloatのイメージに変換すること。詳細はConvertingを参照**

- tf.image.resize_images() # 普通のリサイズ
- tf.image.resize_area() # 各種リサイズ手法毎にメソッドが切られている
以下略



```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import tensorflow as tf
from PIL import Image # 画像表示用
import matplotlib.pyplot as plt # 画像表示用

value = tf.read_file("data/image.jpg")
image = tf.image.decode_jpeg(value, channels=3) # RGB => channels=3
image = tf.image.convert_image_dtype(image, dtype=tf.float32)
image = tf.image.resize_images(image, (200,200), method=tf.image.ResizeMethod.BICUBIC)

with tf.Session() as sess:
    img = sess.run(image)
    plt.imshow(img)
    plt.show()
```
