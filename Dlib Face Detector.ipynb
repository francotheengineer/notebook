```python
import sys
import dlib
from skimage import io


from PIL import Image # 画像表示用
import matplotlib.pyplot as plt # 画像表示用
```

```python
detector = dlib.get_frontal_face_detector()
```

```python
img = io.imread("data/image4.jpg")
```

```python
dets = detector(img, 1)
```

```python
print("Number of faces detected: {}".format(len(dets)))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.imshow(img)

for i, d in enumerate(dets):
    print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
        i, d.left(),  d.top(), d.right(), d.bottom()))
    ax.add_patch(plt.Rectangle((d.left(), d.bottom()), d.right() - d.left(), d.top() - d.bottom(), fill=False, edgecolor="yellow"))

plt.show()
```

```python
from scipy import misc 

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
for i, d in enumerate(dets):
    print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
        i, d.left(),  d.top(), d.right(), d.bottom()))
    imgt = img[d.top():d.bottom(), d.left():d.right(), :]
    imgt = misc.imresize(imgt, (230, 230), 'bicubic')
    plt.imshow(imgt)
    
plt.show()

io.imsave("./data/base_fase.jpg", imgt)
```
