# Procrustes analysis
2つの特徴点の集合を重ね合わせて、最もFitするように特徴点を移動・回転・変形させる手法

```python
from skimage import io
import dlib
import glob

import numpy as np
from scipy import ndimage

from PIL import Image # 画像表示用
import matplotlib.pyplot as plt # 画像表示用
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("./data/shape_predictor_68_face_landmarks.dat")
```

```python
# 基準点を算出(適当にこの画像)
f = "./data/image4.jpg"
img = io.imread(f)
dets = detector(img, 1)
for k, d in enumerate(dets):
    shape = predictor(img, d).parts()
    shape_base = np.moveaxis(np.array([[point.x for point in shape], [point.y for point in shape]], dtype=float), 1, 0)

average = np.mean(shape_base, axis=0)
stdA = np.std(shape_base)

shape_base =(shape_base - average) / stdA
v = shape_base.T

shift = np.concatenate((-average, [0]))

img = ndimage.shift(img, shift)
img = ndimage.zoom(img, np.array([stdA, stdA, 1]))

print(np.shape(img))

plt.imshow(img)
plt.scatter(v[0], v[1], color='yellow')
plt.show()
print(average, stdA)
```

```python
for f in glob.glob("./dlib_data/*.jpg"):
    print("Processing file: {}".format(f))
    img = io.imread(f)
    
    dets = detector(img, 1)
    for  k, d in enumerate(dets):
        # Get the landmarks/parts for the face in box d.
        shape = predictor(img, d).parts()
        shape = np.moveaxis(np.array([[point.x for point in shape], [point.y for point in shape]], dtype=float), 1, 0)

    average = np.mean(shape, axis=0)
    stdB = np.std(shape)

    shape = (shape - average) / stdB
    A = np.dot(shape_base.T, shape)
    U, s, Vt = np.linalg.svd(A)
    V = Vt.T
    T = np.dot(V, U.T)
    
    traceTA = s.sum()
    
    b = 1
    d = 1 - 2 * traceTA
    Z = np.dot(shape, T)
    
#     print(Z)
    vt = Z.T
    plt.scatter(v[0], v[1])
    plt.scatter(vt[0], vt[1], edgecolors="Yellow")
    plt.show()
    
#     print(np.shape(s), np.shape(U), np.shape(V))
#     x = [point.x for point in shape.parts()]
#     y = [point.y for point in shape.parts()]
#     ax_img = plt.subplot(1, 1, 1)
#     ax_img.imshow(img)
#     plt.scatter(x, y, s=5, c='blue')

#     plt.show()
```
